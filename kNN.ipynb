###### Main Euclideian distance will be calculated here

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,query_x,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=dist(query_x,X[i])
        vals.append((d,Y[i]))
    
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
#     c1=0
#     for tup in vals:
#         if(tup[1]==0) :
#             c1 +=1
#     if(c1*2>=k) :
#         print("Class 0")
#     else:
#         print("Class 1")
    newVal=np.unique(vals[:,1],return_counts=True)
#     print(newVal)
    idx=newVal[1].argmax()
    pred=newVal[0][idx]
    return pred
#     return vals
#     print(vals)
```

```python
knn(X,Y,query_point)
```
