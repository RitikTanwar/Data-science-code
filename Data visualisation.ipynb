## Data Visualisation

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

###### Line Plot


```python
x=np.arange(10)
y=x**2
z=2*x+3
print(x)
print(y)
print(z)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z,color='red')
plt.show()
```

```python
plt.plot(x,y,label='Apple')
plt.plot(x,z,color='red',label='Kiwi',linestyle='dashed',marker='o')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price of fruits over time')
plt.legend()  #for labelling curve 
plt.show()
```

###### Scatter plots

```python
plt.scatter(x,y,marker='*',label='Apple',color='red')
plt.scatter(x,z,marker='o',label='Kiwi',color='green')
plt.legend()
plt.show()
```

###### Bar

```python
plt.bar?
```

```python
plt.bar([1,2,3],[10,20,15]) #cur year
plt.bar([1,2,3],[15,10,20])
plt.show()
```

To make these overlapping bar beautiful

```python
x_cor=np.array([1,2,3])**2
plt.bar(x_cor,[10,20,15],width=0.5,label='Current year',tick_label=['Gold','Platinum','Silver'])
plt.bar(x_cor+0.5,[15,10,20],width=0.5,label='Next year',color='orange')
plt.title('Metal Price comparison')
plt.xlabel('Metal')
plt.ylabel('Price')
plt.ylim(0,30)
plt.legend()
plt.show()
```

```python
subjects=['OS','OOPs','DBMS','CN','DSA']
weightage=[20,30,10,10,10]
plt.pie(weightage,labels=subjects,explode=(0,0.1,0,0,0.1),autopct="%1.1f%%",shadow=True)
plt.show()
```

###### Histogram

```python
xsn=np.random.randn(100)
sigma=5
u=60
X=np.round(xsn*sigma+u)
Y=np.round(xsn*8+40)
print(X)
```

```python
plt.style.use('dark_background')
plt.hist(X,alpha=0.7)
plt.hist(Y,alpha=0.7)
plt.xlabel('No. of students')
plt.ylabel('Marks')
plt.show()
```
