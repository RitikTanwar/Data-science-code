```python
from urllib.request import urlopen
```

```python
amazon_url="https://www.amazon.in/"
```

```python
amazon_data=urlopen(amazon_url)
```

```python
amazon_html=amazon_data.read()
amazon_data.close()
```

```python
amazon_html
```

```python
from bs4 import BeautifulSoup as soup
```

```python
amazon_soup=soup(amazon_html,'html.parser')
```

```python
amazon_soup
```

```python
img_links=[]
for item in amazon_soup.findAll('img'):
    link=item['src']
    print(item['src'])
    img_links.append(link)
```

```python
img_links
```

```python
print(len(img_links))
```

```python
img_url=img_links[0]
```

```python
img_url
```

```python
import requests
```

```python
response=requests.get(img_url)
```

```python
print(response.content)
```

```python
with open('amazon.jpg','wb') as f:
    response=requests.get(img_links[0])
    f.write(response.content)
```
