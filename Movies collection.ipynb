```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('movie_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
titles=df.get('movie_title')
print(titles)
```

```python
freq_title={}
```

```python
for t in titles:
    length=len(t)
    print(t,length)
```

```python
for t in titles:
    length=len(t)
    try :
        freq_title[length]+=1
    except:
        freq_title[length]=1
```

```python
freq_title
```

```python
X=np.array(list(freq_title.keys()))
Y=np.array(list(freq_title.values()))
```

```python
plt.style.use('seaborn')
```

```python
plt.scatter(X,Y)
plt.xlabel('Movie Title length')
plt.ylabel('Freq of movies')
plt.legend()
plt.show()
```
