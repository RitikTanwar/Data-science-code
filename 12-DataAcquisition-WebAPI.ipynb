### Web Api

- Application Prog. Interface:- Set of routines running over server
- Connects Application and database
- Speficify Data with URL
- Receive data and parse it
- Api may return data in Json or xml

##### JSON

- JavaScript Object Notation
- Dictionaty like object i.e. key-value pair
- Keys are string value can be anything
- Easy to read

##### XML

- eXtensible Markup Language
- structure similar to XML
- Custom tags
- No style

#### Types of request
- Get
- Post
- Delete
- Put
- Patch

- urllib
- requests

```python
from urllib.request import urlopen
```

```python
url_api='http://api.openweathermap.org/data/2.5/weather?q=London,uk&APPID=17b56b9e9716992c56820c3537854581'
```

```python
url_res=urlopen(url_api)
```

```python
url_data=url_res.read()
print(url_data)
```

```python
print(type(url_data))
```

    We have to convert this data into json

```python
import json
json_data=json.loads(url_data)
```

```python
json_data  #this is dict i.e. key value pair
```
