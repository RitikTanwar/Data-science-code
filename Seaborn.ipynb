```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_style('dark')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) #by default estimator is mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.var )
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=True)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#kind is by-default scatter
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',color='pink')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',color='pink')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hist',color='pink')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='pink')
```

```python
sns.pairplot(tips,height=5)
```

```python
sns.pairplot(tips,height=5,hue='sex')
```
