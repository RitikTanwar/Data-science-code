```python
u=10
sigma=2
val=u+sigma*np.random.randn(1000)
plt.hist(val,100)
plt.show()
```

### Multivariate normal distribution

###### Bivariate

```python
u=np.array([0.0,0.0])
cov=np.array([[1,0.8],[0.8,1]])

u2=np.array([5.0,6.0])
cov2=np.array([[1.3,0.2],[0.2,1.1]])

dist=np.random.multivariate_normal(u,cov,500)
dist2=np.random.multivariate_normal(u2,cov2,500)
# print(dist)
# dist.shape
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
