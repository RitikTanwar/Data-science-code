### Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=0
sigma=1
val=np.random.randn(1000)
```

```python
plt.style.use('seaborn')
plt.hist(val)
plt.show()
```

This is standard normal distribution

```python
u=10
sigma=2
val=u+sigma*np.random.randn(1000)
plt.hist(val,100)
plt.show()
```

```python
#Checking how many values are close to mean
val=np.round(val)
z=np.unique(val,return_counts=True)
print(z)
```

```python
u=10
sigma=2
val=u+sigma*np.random.randn(200)
u2=12
sigma2=1
val2=u2+sigma2*np.random.randn(200)
plt.hist(val,100)
plt.hist(val2,100)
plt.show()
```

```python
x=val
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```
